```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
aprime = pd.read_csv("amazon_prime_titles.csv")
```

```python
aprime['cast'].fillna("Unknown", inplace=True)
```

```python
aprime.info()
```

```python
aprime["director"].fillna("Director Unknown" ,inplace = True)
```

```python
aprime
```

```python
aprime.info()
```

```python
aprime["country"].value_counts()
```

```python
#store_df["Type"].fillna(str(store_df["Type"].mode().values[0]),inplace = True)
aprime["country"].mode()
```

```python
# replacing NaN values by most common country using fillna and mode 
aprime["country"].fillna(aprime["country"].mode()[0],inplace = True)
```

```python
aprime
```

```python
# replacing NaN values by most common rating using fillna and mode
aprime["type"].mode()
```

```python
aprime["type"].value_counts()
```

```python
aprime["rating"].fillna(aprime["rating"].mode()[0],inplace = True)
```

```python
aprime
```

```python
aprime["date_added"].fillna("Unkown" ,inplace = True)
```

```python
aprime
```

```python
aprime.info()
```

```python
aprime["listed_in"].value_counts()
```

```python
aprime["listed_in"].value_counts().nlargest(10)
```

```python
aprime["rating"].unique()
```

```python
top_ten_ratings = aprime["rating"].value_counts().nlargest(10)
top_ten_ratings
```

```python
plt.figure(figsize=(10, 6))
top_ten_ratings.sort_index().plot(kind='bar', color='pink')
plt.title('Rating by total Shows')
plt.xlabel('Top Ten Rating')
plt.ylabel('Total Shows')
```

```python
top_ten_genres = aprime["listed_in"].value_counts().nlargest(10)
top_ten_genres
```

```python
plt.figure(figsize=(10, 6))
top_ten_genres.sort_index().plot(kind='bar', color='purple')
plt.title('Genres by total Shows')
plt.xlabel('Top Ten Genres')
plt.ylabel('Total Shows')
```

```python
aprime["release_year"].value_counts()
```

```python
release_year_count = aprime["release_year"].value_counts()
release_year_count
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(aprime['release_year'],bins=30, kde=True, color='green')
plt.title('Total Shows by Release Year')
plt.xlabel('release_year')
plt.ylabel('Total Shows')
```

```python
aprime["country"].value_counts().nlargest(5)
```

```python
top_five_country = aprime["country"].value_counts().nlargest(5)
top_five_country
```

```python
plt.figure(figsize=(10, 6))
top_five_country.sort_index().plot(kind='bar', color='yellow')
plt.title(' Total Shows by Country')
plt.xlabel('Top five Country')
plt.ylabel('Total Shows')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_ten_genres.values, y=top_ten_genres.index, palette='rocket')
plt.title('Top 10 Genres')
plt.xlabel('Number of Shows')
plt.ylabel('Genre')
```

```python
total_types = aprime["type"].value_counts()
total_types
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='type', data=aprime, palette='magma')
plt.title('Total of Movies and TV Shows')
plt.xlabel('Type')
plt.ylabel('shows')
```
